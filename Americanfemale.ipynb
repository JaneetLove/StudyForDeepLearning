```python
#Agriculture
import pandas as pd
import matplotlib.pyplot as plt
unrate = pd.read_csv('percent-bachelors-degrees-wome.csv')#读取文件
first_t = unrate[0:21]
plt.plot(first_t['Year'],first_t['Agriculture'])
plt.xlabel("Year")#横坐标标签
plt.ylabel("Percentage")#纵坐标标签
plt.title('Agriculture')#图标标题
plt.show()
```

```python
#绘制多个子图
import pandas as pd
import matplotlib.pyplot as plt
unrate = pd.read_csv('percent-bachelors-degrees-wome.csv')#读取文件
first_t = unrate[0:21]
fig = plt.figure()
Architecture = fig.add_subplot(3,2,1)
ArtPerformance = fig.add_subplot(3,2,2)
Business = fig.add_subplot(3,2,3)
Biology = fig.add_subplot(3,2,4)
ComputerScience = fig.add_subplot(3,2,5)
Education = fig.add_subplot(3,2,6)

Architecture.plot(first_t['Year'],first_t['Architecture'])
ArtPerformance.plot(first_t['Year'],first_t['Art and Performance'])
Business.plot(first_t['Year'],first_t['Business'])
Biology.plot(first_t['Year'],first_t['Biology'])
ComputerScience.plot(first_t['Year'],first_t['Computer Science'])
Education.plot(first_t['Year'],first_t['Education'])
plt.show()
```

```python
#一个图中多个曲线
import pandas as pd
import matplotlib.pyplot as plt
unrate = pd.read_csv('percent-bachelors-degrees-wome.csv')#读取文件
first_t = unrate[0:21]
plt.plot(first_t['Year'],first_t['Architecture'],c ='red')
plt.plot(first_t['Year'],first_t['Art and Performance'],c ='orange')
plt.plot(first_t['Year'],first_t['Business'],c ='yellow')
plt.plot(first_t['Year'],first_t['Biology'],c ='green')
plt.plot(first_t['Year'],first_t['Computer Science'],c ='blue')
plt.plot(first_t['Year'],first_t['Education'],c ='purple')

plt.legend()
plt.show()
```
