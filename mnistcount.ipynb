```python
import sys, os
sys.path.append(os.pardir)  
import numpy as np
import pickle
from dataset.mnist import load_mnist
def sigmoid(x):
    return 1/(1+np.exp(-x))
def softmax(a):
    c=np.max(a)
    exp_a=np.exp(a-c)
    sum_exp_a=np.sum(exp_a)
    y=exp_a/sum_exp_a
    return y
#损失函数(均方误差)
def mean_squared_error(y,t):
    return 0.5 * np.sum((y-t)**2)
#损失函数(交叉熵误差)
def cross_entropy_error(y,t):
    delta = 1e-7
    return -np.sum(t * np.log(y + delta))

def get_data():
    (x_train, t_train), (x_test, t_test) = load_mnist(normalize=True, flatten=True, one_hot_label=True)
    return x_test, t_test


def init_network():
    with open("sample_weight.pkl", 'rb') as f:
        network = pickle.load(f)
    return network


def predict(network, x):
    #TODO
    W1,W2,W3=network['W1'],network['W2'],network['W3']
    b1,b2,b3=network['b1'],network['b2'],network['b3']
    #第一层
    a1=np.dot(x,W1)+b1
    z1=sigmoid(a1)
    #第二层
    a2=np.dot(z1,W2)+b2
    z2=sigmoid(a2)
    #第三层
    a3=np.dot(z2,W3)+b3
    y=softmax(a3)
    
    return y


x, t = get_data()
network = init_network()
accuracy_cnt = 0
#TODO
for i in range(20):
    y = predict(network,x[i])
    k = np.argmax(y)
    L = np.argmax(t[i])
    p = mean_squared_error(np.array(y),np.array(t[i]))
    q = cross_entropy_error(np.array(y),np.array(t[i]))
    print("第"+str(i+1)+"个"+" "+"预测值:"+str(k)+" "+"标注值:"+str(L)+" "+"均方误差:"+str(p)+" "+"交叉熵误差:"+str(q))
(x_train, t_train), (x_test, t_test) = load_mnist(normalize=True, flatten=True, one_hot_label=True)
print("前20张batch")
for i in range(20):
    y = predict(network,x_train[i])
    k = np.argmax(y)
    L = np.argmax(t_train[i])
    p = mean_squared_error(np.array(y),np.array(t_train[i]))
    q = cross_entropy_error(np.array(y),np.array(t_train[i]))
    print("第"+str(i+1)+"个"+" "+"预测值:"+str(k)+" "+"标注值:"+str(L)+" "+"均方误差:"+str(p)+" "+"交叉熵误差:"+str(q))
    
```
