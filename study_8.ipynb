```python
import numpy as np
class MulLayer:
    def __init__(self):
        self.x = None
        self.y = None
        
    def forward(self,x,y):
        self.x = x
        self.y = y
        out = x * y
        
        return out
    def backward(self,dout):
        dx = dout * self.y#反转x和y
        dy = dout * self.x
        
        return dx,dy
class AddLayer:
    def __init__(self):
        pass
    
    def forward(self,x,y):
        out = x + y
        return out
    
    def backward(self,dout):
        dx = dout * 1
        dy = dout * 1
        return dx,dy
class Sigmoid:
    def __init__(self):
        self.out = None
        
    def forward(self,x):
        out = 1 / (1+np.exp(-x))
        self.out = out
        
        return out
    def backward(self,dout):
        dx = dout * (1.0-self.out) * self.out
        
        return dx
class mullayer3:
    def __init__(self):
        self.x = None
        self.y = None
        self.z = None

    def forward(self,x,y,z):
        self.x = x
        self.y = y
        self.z = z
        out = x * y * z
        return out
    def backward(self,dout):
        dx = dout * self.y * self.z 
        dy = dout * self.x * self.z
        dz = dout * self.x * self.y
        return dx,dy,dz

x1,x2 = 0.02,0.03

#layer
add_xy1_layer = AddLayer()
add_xy2_layer = AddLayer()
mul_xy_layer = MulLayer()
mul_abc_layer = mullayer3()
sigmoid_layer = Sigmoid()
#forward
a = add_xy1_layer.forward(x1,x2)
b = add_xy2_layer.forward(x1,x2)
c = mul_xy_layer.forward(x1,x2)
abc = mul_abc_layer.forward(a,b,c)
end3 = sigmoid_layer.forward(abc)
#backward
dend3 = 2
dabc = sigmoid_layer.backward(dend3)
da,db,dc = mul_abc_layer.backward(dabc)
x1,y1 = add_xy1_layer.backward(da)
x2,y2 = add_xy2_layer.backward(db)
x3,y3 = mul_xy_layer.backward(dc)
print("正向传播")
print("y = ",end3)
print("第一层正向输出：",a,b,c)
print("第二层正向输出：",abc)
print("第三层正向输出：",end3)

print("反向传播")
print("反馈误差 = ",dend3)
print("第一层反向输出：",x1+x2+x3,y1+y2+y3)
print("第二层反向向输出：",da,db,dc)
print("第三层反向输出：",dabc)
```
